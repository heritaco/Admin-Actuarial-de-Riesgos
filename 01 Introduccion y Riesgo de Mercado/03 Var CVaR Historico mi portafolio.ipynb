```python
import numpy as np
import pandas as pd
import plotly.express as px
from src import style
style.set_plotly_defaults()
```

```python
csv_path = "data/Tech_industry_prices-2005.csv"
df = pd.read_csv(csv_path, parse_dates=["Date"]).set_index("Date").sort_index()
df
```

```python
rets = df / df.shift(1) -1
rets
```

```python
w_lowv = pd.Series({
    "ROP": 2.070828e-01,
    "IBM": 2.040559e-01,
    "MSI": 1.438235e-01,
    "CSCO": 1.196090e-01,
    "VRSN": 9.148111e-02,
    "GEN": 6.869088e-02,
    "TDY": 5.082798e-02,
    "MSFT": 3.704157e-02,
    "GLW": 2.975414e-02,
    "AKAM": 2.203776e-02,
    "FSLR": 1.881911e-02,
    "GDDY": 6.776196e-03
},  name="weight",
)

w_maxsh = pd.Series({
    "IBM": 3.723745e-01,
    "APH": 3.115756e-01,
    "NVDA": 8.584984e-02,
    "MSI": 7.056857e-02,
    "AVGO": 6.649925e-02,
    "STX": 4.078228e-02,
    "JBL": 3.319984e-02,
    "FSLR": 1.100151e-02,
    "ANET": 7.687008e-03,
    "APP": 4.616639e-04
}, name = "weight",)

# sanity checks
if not np.isclose(w_lowv.sum(), 1.0):
    raise ValueError(f"Weights must sum to 1. Current sum={w_lowv.sum()}")

port_lowv = (rets * w_lowv).sum(axis=1).rename("port_ret")
port_maxsh = (rets * w_maxsh).sum(axis=1).rename("port_ret")

df_portlowv = pd.DataFrame({"port_ret": port_lowv})
df_portmaxsh = pd.DataFrame({"port_ret": port_maxsh})   
```

```python
from plotly.subplots import make_subplots

import plotly.graph_objects as go

# Create subplots with 1 row and 2 columns, both as pie charts
fig = make_subplots(
    rows=1, cols=2,
    specs=[[{'type': 'pie'}, {'type': 'pie'}]],
    subplot_titles=["Weights del portafolio de baja volatilidad", "Weights del portafolio de max sharpe"]
)

# Add the first pie chart for w_lowv
fig.add_trace(
    go.Pie(
        labels=w_lowv.index,
        values=w_lowv.values,
        name="Low Volatility"
    ),
    row=1, col=1
)

# Add the second pie chart for w_maxsh
fig.add_trace(
    go.Pie(
        labels=w_maxsh.index,
        values=w_maxsh.values,
        name="Max Sharpe"
    ),
    row=1, col=2
)

# Update layout if needed
fig.update_layout(title_text="Visualización de los pesos de los portafolios")
# move the title up
fig.update_layout(title_y=1)

fig.show()
```

```python
# visualizacion de los pesos de los portafolios
fig = px.pie(
    names=w_maxsh.index,
    values=w_maxsh.values,
    title="Weights del portafolio de max sharpe",
)
fig.show()
```

# Visualizaciones

```python
rlowv = df_risk_lowv["Returns"].astype(float).dropna()
rmaxs = df_risk_maxsh["Returns"].astype(float).dropna()

cumlowv = (1.0 + rlowv).cumprod().rename("cum_$1 lowv")
cummaxs = (1.0 + rmaxs).cumprod().rename("cum_$1")

temp_df = pd.DataFrame({'Portafolio Low Volatility': cumlowv, 'Portafolio Max Sharpe': cummaxs})

fig = px.line(
    temp_df,
    x=temp_df.index,
    y=['Portafolio Low Volatility', 'Portafolio Max Sharpe'],
    title="Evolución del valor de $1 invertido en el portafolio",
    labels={"x": "Date", "y": "Value"},
)
fig.show()
```

```python
risk_col = df_risk_lowv.columns[1]
cvar_col = df_risk_lowv.columns[2]
var10d_col = df_risk_lowv.columns[3]
capital10d = df_risk_lowv.columns[4]
tmp = df_risk_lowv * -1
fig = px.line(
     tmp,
     x=tmp.index,
     y=[df_risk_lowv["Returns"], tmp[cvar_col], tmp[var10d_col], tmp[capital10d]],
     labels={"value": "Return", "variable": "Series"},
     title=f"Medidas de riesgo para el portafolio de baja volatilidad",
)


fig.show()
```

```python
risk_col = df_risk_maxsh.columns[1]
cvar_col = df_risk_maxsh.columns[2]
var10d_col = df_risk_maxsh.columns[3]
capital10d = df_risk_maxsh.columns[4]
tmp2 = df_risk_maxsh * -1
fig = px.line(
     tmp2,
     x=tmp2.index,
     y=[df_risk_maxsh["Returns"], tmp2[cvar_col], tmp2[var10d_col], tmp2[capital10d]],
     labels={"value": "Return", "variable": "Series"},
     title=f"Medidas de riesgo para el portafolio de max sharpe",
)


fig.show()
```
